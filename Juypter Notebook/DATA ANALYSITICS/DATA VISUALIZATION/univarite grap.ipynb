```python
import pandas as pd
import seaborn as sns
```

```python
fd = pd.read_csv("AB_NYC_2019.csv")
```

```python
fd.head()
```

## 1.Univarite Categorical

### Bar Graph

```python
sns.countplot(fd,x= "neighbourhood_group")
```

```python
sns.countplot(fd,x= "room_type")
```

```python
sns.countplot(data =fd, x = "neighbourhood_group",hue = "room_type")
```

```python
sns.countplot(data =fd, x = "room_type",hue = "neighbourhood_group")
```

### Pie Chart

```python
fd["neighbourhood_group"].value_counts().plot(kind = "pie",autopct = "%.3f")
```

```python
fd["room_type"].value_counts().plot(kind = "pie",autopct = "%.3f")
```
