```python
import pandas as pd
import seaborn as sns 
```

```python
ff = sns.get_dataset_names()
```

```python
fd= sns.load_dataset("tips")
```

```python
fd.head()
```

## Categorical

```python
sns.barplot(data = fd , y="total_bill", x = "sex",estimator = "sum")
```

```python
sns.boxplot(data = fd, x = "sex",y = "tip",hue = "smoker")
```

```python
sns.barplot(data = fd , y="total_bill", x = "sex", hue= "smoker" , estimator = "sum")
```

```python
sns.scatterplot(data = fd , x = "total_bill",y = "tip",hue = "sex",style = "smoker")
```

```python
sns.violinplot(data = fd, x = "sex" , y = "tip")
```

```python
sns.violinplot(data = fd, x = "sex" , y = "total_bill",hue = "smoker")
```
