## 3.1 Numeric data

```python
fd3= fd.copy()
```

```python
fd3.head(2)
```

```python
fd3.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
imputer = SimpleImputer(missing_values = np.nan,strategy = "mean")
```

```python
fd3.loc[:,"reviews_per_month":"reviews_per_month"] = imputer.fit_transform(fd3.loc[:,"reviews_per_month":"reviews_per_month"])
```

```python
fd3.isna().sum()
```

## 3.2 categorical

```python
fd4 = fd.copy()
```

```python
fd4.sample(10)
```

```python
fd4.isna().sum()
```

```python
imputer = SimpleImputer(missing_values = np.nan,strategy = "most_frequent")
fd4.loc[:,"host_name":"host_name"] = imputer.fit_transform(fd4.loc[:,"host_name":"host_name"])
```

```python
fd4.isna().sum()
```
